# Explicación primer paso 

Se guardaron en diccionario las comorbilidades y las fechas se guardaron como datos, se crean dos diccionarios anidados que tendría que tener una fecha como llave y dentro de cada fecha, un hospitalizado y un no hospitalizado, ocupamos columna.pop(0), para que no nos tome la primera línea de (cormobilidad y hospitalización), ya que no es lo que queremos utilizar para el grafico, al repetirse mis datos en el CSV, es ideal crear diccionario que me permita que solo me tome un dato, es decir,  al estar la llave esto me permitirá no agregar una nueva, las fechas serán guardada como arreglo, el cual se utilizara un contador para moverse entre ellas, para luego inicializar mis dos diccionarios anidados de pacientes_si, y pacientes_no, refiriéndose a los que fueron hospitalizados y a los que no.


```python
import csv 
```

```python
#Cormobilidades se guardara como diccionario.
cormobilidades={}
#Las fechas seguardaran como datos. 
fechas=[]



#Esto refiere a pacientes si hospitalizados y pacientes no hospitalizados. 
pacientes_no={}
pacientes_si={}


with open ('data/Comorbilidad.csv') as cormobilidad:
    linea = csv.reader(cormobilidad, delimiter=',')
    bandera=1
    
    for columna in linea:
        if bandera==1:
            #Esto permitirá no recorrer en mi CSV Cormobilidad ni Hospitalización. 
            columna.pop(0)
            columna.pop(0)
            #Se guardara como arreglo en fechas, el cual ya fue creado. 
            fechas=columna
            bandera=0
        else: 
            #Contador permitira recorrer fechas. 
            contador=2
            
            cormobilidades[columna[0]]=1 
            for fecha in fechas:
                if 'NO' in columna[1]:
                    try: 
                        pacientes_no[fecha][columna[0]]=columna[contador]
                    except KeyError:
                        #Inicializacion de diccionario anidado.
                        pacientes_no[fecha]={}
                        pacientes_no[fecha][columna[0]]={}
                        pacientes_no[fecha][columna[0]]=columna[contador]
                else:
                    try:
                        pacientes_si[fecha][columna[0]]=columna[contador]
                    except KeyError:
                        #Inicializacion de diccionario anidado.
                        pacientes_si[fecha]={}
                        pacientes_si[fecha][columna[0]]={}
                        pacientes_si[fecha][columna[0]]=columna[contador]
                        
                    
                contador+=1 
                #Contador que permite recorrer mi CSV
                
                
                              
        
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

# Explicación séptimo paso 

Ocupamos un espacio de figsize de (100,0) al ser un gráfico muy extenso quede legible, fijamos que nuestros datos están guardados en una cadena de String, lo cual nos perjudica si queremos grafiar, utilizamos otro carácter que es el float para llevarlo a una cadena numérica y nos permita poder graficar, para que nuestro grafico mejore su diseño, utilizamos fontsize, comando que nos permitirá agrandar las letras de nuestro grafico para que quede un diseño más agradable a la vista. Anidado a esto se generará primero un gráfico con los pacientes que no fueron hospitalizados, y luego en la siguiente línea se hará con los que si fueron hospitalizados, con esto damos por terminado los primeros gráficos, también llamaremos al arreglo enX, guardado en una llave donde se encuentran nuestras cormobilidades en el tiempo.   



```python
fig,ax=plt.subplots(figsize=(100,10))

ejeX= np.arange(len(fechas))
#Se genera un diccionario con las llaves para cada comorbilidad en el tiempo. 
enX={}
for comorb in cormobilidades:
    enX[comorb]=[]
    
for fecha in fechas: 
    for comorb in cormobilidades: 
        #El float se utiliza para pasar mis datos guardados como String a float.
        enX[comorb].append(float(pacientes_no[fecha][comorb]))
            
for comorb in cormobilidades: 
    ax.plot(enX[comorb],label=comorb)

ax.legend()
ax.set_xticks(ejeX)
ax.set_xticklabels(fechas, rotation=90, fontsize=20) #Utilizamos fontsize para hacer mas legible la letra de nuestro gráfico.
ax.set_title("Pacientes con Covid que NO fueron hospitalizados por Cormobilidades", fontsize=40)
ax.set_xlabel("Días", fontsize=40)
ax.set_ylabel("Cantidad de pacientes", fontsize=40)
```
